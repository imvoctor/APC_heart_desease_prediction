```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import scikitplot as skplt
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_curve, auc
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
```

```python
dataset = pd.read_csv('heart_desease_data.csv')
```

```python
dataset.head()
```

Las dadas de la base de datos , nos indican la informacion médica de diferentes personas, en concreto nos dan 14 parametros.
Creo que nuestro atributo objetivo serà el target ya que podemos ver que és la variable que nos marca si un paciente tendrá o no una enfermedad cardiaca.

- age - age in years
- sex - sex (1 = male; 0 = female)
- cp - chest pain type (1 = typical angina; 2 = atypical angina; 3 = non-anginal pain; 4 = asymptomatic)
- trestbps - resting blood pressure (in mm Hg on admission to the hospital)
- chol - serum cholestoral in mg/dl
- fbs - fasting blood sugar > 120 mg/dl (1 = true; 0 = false)
- restecg - resting electrocardiographic results (0 = normal; 1 = having ST-T; 2 = hypertrophy)
- thalach - maximum heart rate achieved
- exang - exercise induced angina (1 = yes; 0 = no)
- oldpeak - ST depression induced by exercise relative to rest
- slope - the slope of the peak exercise ST segment (1 = upsloping; 2 = flat; 3 = downsloping)
- ca - number of major vessels (0-3) colored by flourosopy
- thal - 3 = normal; 6 = fixed defect; 7 = reversable defect
- target - the predicted attribute - (0 = absence, 0< = presence)

Ahora separamos la variable objetivo del dataset y dejamos el dataset con las variables con mas relación.

```python
attributes = ['slope', 'thalach', 'cp', 'thal', 'ca', 'oldpeak', 'exang', 'age', 'sex']
dataset_y = dataset[attributes]
y = dataset['target']
dataset_y.head()
```

Podemos ver el datasat reducido con solo los atributos que nos interesan, y si la variable objetivo para poder partir los datos mas adelantem.

# Preparacion de datos para entrenamiento de modelos

```python
dataset_y = preprocessing.normalize(dataset_y)
print(dataset_y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(dataset_y, y, test_size = 0.20)
```

## Model selection

### Random Forest.

Utilizamos la funcion de randomforest de sklearn para generar un bosque de 500 arboles.
Con el modelo hecho lo utilizamos para entrenar los datos partidos anteriormente, predecimos con el test y luego calculamos cuanta accuracy hemos tenido con este modelo.

```python
random = RandomForestClassifier(n_estimators=500, random_state = 123)
random_final = random.fit(x_train, y_train)
y_pred = random_final.predict(x_test)
prob = random_final.predict_proba(x_test)

accuracy = round(accuracy_score(y_test, y_pred)*100, 2)
model_accuracy = round(random_final.score(x_train, y_train)*100,2)
model_accuracy_test = round(random_final.score(x_test, y_test)*100,2)


print('Accuracy:', accuracy, '%')
print('Model accuracy train:', model_accuracy, '%')
print('Model accuracy test:', model_accuracy, '%')
```

Hacemos una roc curve para mostrar mejor como de buena es la predicción, tendremos dos classes: 0 y 1, absence o presence de
enfermedad cardiaca. 
A priori como mas cerca del 1 esté el area de la clase mejor.

```python
fpr = {}
tpr = {}
roc_auc = {}


for i in range(2):
    fpr[i], tpr[i],_  = roc_curve(y_test == i, prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure()
for i in range(2):
    plt.plot(fpr[i], tpr[i], label='Class {0} (area = {1:0.2f})' ''.format(i, roc_auc[i]))
plt.legend()
```

Ahora hacemos una confussion matrix para ver los valors de TP TN FP FN, lo que nos interesa en este caso, es que no haya FN ya que no queremos que se nos pasen personas que realmente estan enfermas. Por lo que nos interesa tambien tener una presicion alta.

```python
confusion_matrix(y_test, y_pred)
plot_confusion_matrix(random_final, x_test, y_test, cmap=plt.cm.Blues)
plt.show()
print(classification_report(y_test, y_pred))
```

Podemos ver que con este modelo tenemos un rango de FN muy bajo, podemos apreciarlo en el quadrado de abajo a la izquierda,
muy pocos pacientes han sido diagnosticado que no tendran ninguna enfermedad cuando realmente si tendrian.
Tambien podemos ver que hay bastante pocos FP, es decir que no diagnosticamos a muchos pacientes sanos como enfermos o possibles enfermos. Tambien vemos en la ultima tabla que tenemos una f1-escore buena al igual que la accuracy.

### SVM

Implementemos el SVM para poder comparar si el random forest es el mejor modelo para classificar los datos.

```python
from sklearn import svm
linear_svm= svm.SVC(kernel='linear', probability=True)
linear_svm.fit(x_train,y_train)
result_linear = linear_svm.predict(x_test)
probs_linear = linear_svm.predict_proba(x_test)
accuracy_linear = round(accuracy_score(y_test, result_linear)*100, 2)

rbf_svm= svm.SVC(kernel='rbf', probability=True)
rbf_svm.fit(x_train,y_train)
result_rbf = rbf_svm.predict(x_test)
probs_rbf= rbf_svm.predict_proba(x_test)
accuracy_rbf = round(accuracy_score(y_test, result_rbf)*100, 2)

print('Accurac linear:', accuracy_linear, '%')
print('Accuracy rbf: ', accuracy_rbf, '%')
```

Podemos ver que el SVM da unos valores muy inferiores al Random forest en cuanto a acuracy, aun asi hacemos la roc curve para verlo mejor.

```python
fpt = {}
tpt = {}
roc_auc_svm = {}


for i in range(2):
    fpt[i], tpt[i],_  = roc_curve(y_test == i, probs_linear[:, i])
    roc_auc_svm[i] = auc(fpt[i], tpt[i])

plt.figure()
for i in range(2):
    plt.plot(fpt[i], tpt[i], label='Class {0} (area = {1:0.2f})' ''.format(i, roc_auc_svm[i]))
plt.legend()
```

```python
confusion_matrix(y_test, result_linear)
plot_confusion_matrix(linear_svm, x_test, y_test, cmap=plt.cm.Blues)
plt.show()
print(classification_report(y_test, result_linear))
```

Viendo la confussion matrix vemos q nos dan unos valores de f1 score bastante bajos. Nos interesa que sea al rebes.
De momento el RF és el mejor modelo.

## Red Neuronal con tensorflow

Implementamos una red neuronal con tensorflow para compararlo con el RF.
Lo primero q hacemos es importar todo lo necessario, en este caso utilizaremos la red neuronal de tensorflow.
Así que primero creamos el modelo con sequential, luego entrenamos el modelo con los datos de train y predecimos con el test.
Entrenamos con 500 epocas ya que asi puede aprender, más de 500 epocas no nos serviria de nada ya que no sube la accuracy.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
import numpy as np

```

```python
modelo = tf.keras.Sequential()
modelo.add(Dense(128,input_shape=(9,),activation='relu'))
modelo.add(Dense(1,activation='sigmoid'))
     

modelo.compile(
    loss='binary_crossentropy', 
    optimizer='adam', 
    metrics=['accuracy'] 
)
     

modelo_fit = modelo.fit(x_train,y_train,epochs=500)

prediction = modelo.predict(x_test)
```

```python
modelo.evaluate(x_test, y_test)
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Accuracy")
plt.plot(modelo_fit.history["accuracy"])
```

Comparando la red neuronal con el RF, podriamos escoger cualquira de los dos ya que nos da unos valores parecidos. Igualmente sabemos seguro que el RF tiene muy pocos FN y pocos FP, así que me decantaria por el RF.

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(x_train, y_train)
result_tree = decision_tree.predict(x_test)
acc_decision_tree_train = round(decision_tree.score(x_train, y_train) * 100, 2)
acc_decision_tree_test = round(decision_tree.score(x_test, y_test) * 100, 2)
accuracy = round(accuracy_score(y_test, result_tree)*100, 2)
print('Accuracy train:', acc_decision_tree_train)
print('Accuracy test: ',acc_decision_tree_train)
print('Acurracy: ', accuracy)
```

```python
confusion_matrix(y_test, result_tree)
plot_confusion_matrix(decision_tree, x_test, y_test, cmap=plt.cm.Blues)
plt.show()
print(classification_report(y_test, result_tree))
```

En comparación con el RF nos da unos valores notablemente mas bajos. Nos seguiriamos quedando con el RF.
Y como segunda opción, aun que tiene valores similares al modelo svm, nos quedariamos con el Decission Tree ya que tiende a tener una F1 score mas elevada.
